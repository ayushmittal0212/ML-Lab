```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X=[171,151,124,134,156]
y=[80,60,45,50,65]
```

```python
def findCoeff(X,y):
    X_mean=np.mean(X)
    y_mean=np.mean(y)
    den=0
    num=0
    for i in range(len(X)):
        num+=((X[i]-X_mean)*(y[i]-y_mean))
        den+=(X[i]-X_mean)**2
    m=num/den
    c=y_mean-(m*X_mean)
    return m,c
```

```python
m,c=findCoeff(X,y)
print(m,c)
```

```python
height=130
predicted=height*m+c
print(predicted)
```

```python
plt.scatter(X,y)
plt.plot(X,y)
```

```python
def plot_line(X,y,m,c):   
    max_X=np.max(X)+100  
    min_X=np.min(X)-100
    X_plot=np.linspace(min_X,max_X,100)
    y_predicted=m*X_plot+np.repeat(c,100)
    plt.scatter(X,y,color="blue",label="Scatter Points")
    plt.plot(X_plot,y_predicted,color="red",label="Simple Linear Regression")
    plt.legend()
    plt.show()
plot_line(X,y,m,c)
```

```python
data=pd.read_csv('..\Dataset\headbrain.csv')
X2=data['Head Size(cm^3)'].values
y2=data['Brain Weight(grams)'].values
m,c=findCoeff(X2,y2)
plot_line(X2,y2,m,c)
```

```python
rmse=0
for i in range(len(X2)):
    y_pred=c+m*X2[i]
    rmse+=((y2[i]-y_pred)**2)
rmse=np.sqrt(rmse/len(X2))
print(rmse)
```

### Linear Regression using SKLearn

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_csv('..\Dataset\headbrain.csv')
X2=data['Head Size(cm^3)'].values
m=len(X2)
X2=X2.reshape((m,1))
y2=data['Brain Weight(grams)'].values
```

```python
#Model Initialisation
reg=LinearRegression()
#Data Fitting
reg=reg.fit(X2,y2)
#Y Prediction
y_pred=reg.predict(X2)

print(reg.intercept_)
print(reg.coef_)
```

```python
#Model Evaluation
rmse=np.sqrt(mean_squared_error(y2,y_pred))
r2=reg.score(X2,y2)
print(rmse)
print(r2)
```

```python
data2=pd.read_csv('../dataset/FuelConsumption.csv')
data2.head(20)
X3=data2[["ENGINESIZE","CYLINDERS","FUELCONSUMPTION_CITY"]].values
y3=data2["CO2EMISSIONS"].values
```

```python
#Model Initialisation
reg=LinearRegression()
#Data Fitting
reg=reg.fit(X3,y3)
#Y Prediction
y_pred=reg.predict(X3)
plot_line(X3,y3,reg.coef_,reg.intercept_[0])
print(reg.intercept_)
print(reg.coef_)
```

```python
#Model Evaluation
rmse=np.sqrt(mean_squared_error(y3,y_pred))
r2=reg.score(X3,y3)
print(rmse)
print(r2)
```
